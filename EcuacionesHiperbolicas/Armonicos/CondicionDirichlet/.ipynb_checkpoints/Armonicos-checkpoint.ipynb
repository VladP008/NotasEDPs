#### Código para visualizar el fenómeno
- El objetivo es apreciar la diferencia entre los distintos armónicos y ver su desarrollo en tiempo

```python
# Se utilizan las siguientes bibliotecas:
import matplotlib.pyplot as plt    # Para las gráficas
import numpy as np                 # Para el uso de arrays
import matplotlib.animation as animation    # Para la animación
from IPython.display import HTML            # Para poder visualizar la animación en el notebook


# La siguiente instrucción se utiliza para habilitar el modo interactivo en el notebook
%matplotlib notebook

# Parámetros del problema
c = 0.12                  # Velocidad de onda
l = np.pi               # Longitud de la cuerda en este caso: π    
n = 2000                  # Número de puntos para la discretización del dominio espacial (por cada intervalo)
x = np.linspace(0,l, n + 1) # Puntos en el intervalo [-2l,-l]
tmax = 60                      # Tiempo máximo de la animación
t = np.linspace(0, tmax, 50)   # Vector del tiempo

# Modos a mostrar
n1 = 1
n2 = 2
n3 = 4
n4 = 6

# Definición de la función armonico
def armonico(x,n):
    """Función que regresa los valores que corresponden al 
       armónico n
    ----------
    x : array
        Valores del dominio

    n : array
        Valor correspondiente al armónico n

    Returns
    -------
    : array 
        Valores correspondientes al armónico n 
    """

    return np.sin(n*x)

# Graficamos la solución u(x,t) solo para mostrar el desarrollo del armónio en tiempo
def u(n,x,t):
    """Función que regresa los valores de la solución
    ----------
    x : array
        Valores del dominio

    n : array
        Valor correspondiente al armónico n

    t: Valor en el tiempo 
    Returns
    -------
    : array 
        Valores correspondientes al armónico n que varia en tiempo
    """
    
    y = np.sin(n*x)*(np.cos(c*n*t))
    return y
    
# Se crea la figura, se crean los objetos a emplear y se fijan parámetros
fig = plt.figure()
gs = fig.add_gridspec(2, 2, hspace=0.15, wspace=0)
(ax1, ax2), (ax3, ax4) = gs.subplots(sharex='col', sharey='row')
line1, = ax1.plot(x, u(n1,x, t[0]), color='blue', linewidth = 1) # Primera imagen
line2, = ax2.plot(x, u(n1,x, t[0]), color='green', linewidth = 1) # Primera imagen
line3, = ax3.plot(x, u(n3,x, t[0]), color='red', linewidth = 1) # Primera imagen
line4, = ax4.plot(x, u(n4,x, t[0]), color='orange', linewidth = 1) # Primera imagen

# Ajustes de los ejes, etiquetas y título
fig.suptitle('Armónicos')
ax1.set_title(f'{n1}°',fontsize=10)
ax2.set_title(f'{n2}°',fontsize=10)
ax3.set_title(f'{n3}°',fontsize=10)
ax4.set_title(f'{n4}°',fontsize=10)

# Se retiran las etiquetas de ejes y escalas
ax1.axes.set_axis_off()
ax2.axes.set_axis_off()
ax3.axes.set_axis_off()
ax4.axes.set_axis_off()

# Se ajustan los tamaños de los ejes 
ax1.set_ylim([-1.5, 1.5]) 
ax3.set_ylim([-1.5, 1.5]) 

# Función de actualización de cada frame
def update(frame):
    """Función que regresa el valor de la solución para cada frame de la animación
    
    Parametros
    ----------
    frame : int
        Indice del arreglo del tiempo

    """
    
    y1 = u(n1,x, t[frame])
    y2 = u(n2,x, t[frame])
    y3 = u(n3,x, t[frame])
    y4 = u(n4,x, t[frame])
    line1.set_ydata(y1)
    line2.set_ydata(y2)
    line3.set_ydata(y3)
    line4.set_ydata(y4)

# Crear la animación
animacion = animation.FuncAnimation(fig, update, frames=len(t), interval=200)

# Mostrar la animación en el notebook
HTML(animacion.to_jshtml())

# Se guarda en formato gif
# animacion.save(filename="Armonicos_DB_ejemplo2.gif", writer="pillow")
```
