### El Laplaciano en coordenadas rectangulares

Se tiene el siguiente problema 
\begin{equation}
   \left\{
      \begin{aligned}
        \Delta u(x,y) &= 0 \quad 0<x<\pi, 0<y<1, \\
        u(0,y) &= 0 \quad 0<y<1,\\
        u(\pi,y) &= 0 \quad 0<y<1,\\
        u(x,0) &= f_0(x) \quad 0<x<\pi, \\
        u(x,1) &= f_1(x) \quad 0<x<\pi. 
      \end{aligned}
    \right.
\end{equation}


La solución de este problema está dada por 
\begin{equation*}
    \boxed{u(x, y) = \sum_{n = 1}^{\infty}\Big(  \frac{\sinh n(1-y)}{\sinh n}A_n  + \frac{\sinh ny }{\sinh n}B_n\Big)\sin nx}
\end{equation*}
Donde los coeficientes $A_n$ y $B_n$ se obtienen de la serie de Fourier en senos de las funciones $f_0$ y $f_1$, respectivamente. Es decir, 
$$f_0(x) = \sum_{n = 1}^{\infty} A_m \sin mx \quad y \quad f_1(x) = \sum_{n = 1}^{\infty} B_m \sin mx$$

```python
# Se utilizan las siguientes bibliotecas:
import matplotlib.pyplot as plt    # Para las gráficas
import numpy as np                 # Para el uso de arrays
from IPython.display import HTML            # Para poder visualizar la animación en el notebook

# La siguiente instrucción se utiliza para habilitar el modo interactivo en el notebook
%matplotlib inline

# Parámetros del problema
N = 2000                      # Número de puntos para la discretización del dominio espacial
x = np.linspace(0,np.pi, N + 1)   # Puntos en el intervalo [0,π]
y = np.linspace(0,1,N + 1 )   # Puntos en el intervalo [0,1]
X,Y = np.meshgrid(x,y)   # Puntos de [0,π]x[0,1]
U = np.zeros((N+1,N+1))  # Matriz que guardará la solución para cada (x,y)
N_s = 5    # Número de términos para la serie 

def coeff_f0(n):
    """Función que regresa el n-ésimo coeficiente de Fourier de la serie
    de en senos de la función f_0.

    Parametros
    ----------
    n: int
        Número natural mayor o igual a 1.
    
    Returns
    -------
    coeff: float 
        El n-ésimo coeficiente de Fourier de f_0.
    """
    ## La función f_0(x) = 100 sen(x)
    if n == 1:
        coeff = 100
    else:
        coeff = 0
    return coeff

def coeff_f1(n):
    """Función que regresa el n-ésimo coeficiente de Fourier de la serie
    de en senos de la función f_1.

    Parametros
    ----------
    n: int
        Número natural mayor o igual a 1.
    
    Returns
    -------
    coeff: float 
        El n-ésimo coeficiente de Fourier de f_1.
    """
    ## La función f_1 (x) = 0 si 0<x<π/2 ; 100 si π/2<x<π 
    coeff = 100*(2/(n*np.pi))*(np.cos(n*np.pi/2)+(-1)**(n+1))
    return coeff

def sol():
    """Función que regresa el valor u(x,y) hasta un número determinado de terminos N_s

    
    Returns
    -------
    sol: array 
        Solución u(x,y)
    """
    sol = np.zeros((N+1,N+1))
    for m in range(1,N_s):
        sol = ((np.sinh(1-Y)/np.sinh(m))*coeff_f0(m) + (np.sinh(m*Y)/np.sinh(m))*coeff_f1(m))*np.sin(m*X) + sol
    return sol

U = sol()

# Se crea la figura y se fijan parámetros
fig,ax= plt.subplots()
mapa_calor = ax.imshow(U, aspect='auto', cmap='plasma',origin='lower', extent=[0, np.pi,0, 1],vmax=100, vmin=0)

# Añadir una barra de color
calor_bar = plt.colorbar(mapa_calor)
calor_bar.set_label('Temperatura')

# Ajustes de los ejes, etiquetas y título
ax.set_title('Placa rectangular')
ax.grid(color='grey', linestyle='-', linewidth=0.1)
ax.set_ylim([-0.05, 1.05])      # Se fijan los límites del eje y
ax.set_xlim([-0.05, np.pi+0.05]) # Se fijan los límites del eje x
#ax.set_xlabel('x')
#ax.set_xlabel('y')

plt.show()  #Muestra la figura
#plt.savefig("Problema1.PlacaRectangular.3.png")


```
