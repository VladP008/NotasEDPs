#### Ejemplo 2.4. Consideremos el siguiente problema de tránsito con velocidad dependiente del espacio:
\begin{equation}
      \left\{
        \begin{aligned}
          \rho_{t} + (\gamma x \rho)_{x} &= 0  && x \in \mathbb{R}, t>0, \\
          \rho(x,0) &= f(x),
        \end{aligned}
      \right.
\end{equation}
donde $\gamma > 0$ es una constante.

Como se vió en el curso, para cada $\xi\in\mathbb{R}$ la solución es: 
\begin{equation}
 x(t)=\xi e^{\gamma t} ,\quad t>0  
\end{equation}

#### Código de python para visualizar las curvas

```python
# Se utilizan las siguientes bibliotecas:
import matplotlib.pyplot as plt    # Para las gráficas
import numpy as np                 # Para el uso de arrays
import matplotlib.animation as animation    # Para la animación
from IPython.display import HTML            # Para poder visualizar la animación en el notebook

# La siguiente instruccion se utiliza para habilitar el modo interactivo en el notebook
%matplotlib notebook

# Parámetros del problema
# NOTA: Es posible modificar estos parámetros
gamma = 0.05                      # Constante gamma del problema
xi = (-1.0, -0.5, 0, 0.5, 1.0)    # Puntos xi en el eje x (es posible modificar el número)
tmax = 30                         # Tiempo máximo de la animación
t = np.linspace(0, tmax, 50)      # Vector del tiempo

# Se define la ecuación para la curva características de cada punto xi 
x0 = xi[0] * np.exp(gamma * t)    # Punto xi = -1.0
x1 = xi[1] * np.exp(gamma * t)    # Punto xi = -0.5
x2 = xi[2] * np.exp(gamma * t)    # Punto xi = 0.0
x3 = xi[3] * np.exp(gamma * t)    # Punto xi = 0.5
x4 = xi[4] * np.exp(gamma * t)    # Punto xi = 1.0

# Se crea la figura y se fijan parámetros
fig, ax = plt.subplots()
# Se crean lineas para cada punto xi 
line0, = ax.plot([], [], label=f'ξ0 = {xi[0]}') 
line1, = ax.plot([], [], label=f'ξ1 = {xi[1]}') 
line2, = ax.plot([], [], label=f'ξ2 = {xi[2]}')
line3, = ax.plot([], [], label=f'ξ3 = {xi[3]}')
line4, = ax.plot([], [], label=f'ξ4 = {xi[4]}')

# Para mostrar el tiempo en la animación
text = ax.text(0, 28.0, f't= {t[0]}', fontsize=10, color='black', ha='center', va='center',bbox=dict(facecolor='pink', alpha=0.5))  
# Ajustes de los ejes, etiquetas y titulo
ax.set_title(r'Curvas características $x(t)=\xi_i e^{\gamma t} $')
ax.set_ylim([t[0], tmax])    # Se fijan los limites del eje y (ejet del tiempo t)
ax.set_xlim([min(xi) * np.exp(gamma * tmax), max(xi) * np.exp(gamma * tmax)]) # Se fijan los limites del eje x (x)
ax.set_ylabel('t')
ax.set_xlabel('x')
ax.legend()
ax.grid(color='grey', linestyle='-', linewidth=0.1)

# Función de inicialización  
def init():
    line0.set_data([], [])
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    line4.set_data([], [])
    return (line0, line1, line2, line3, line4)

# Función que actualiza cada frame de la animación 
def update(frame):
    xdata = x0[:frame]
    tdata = t[:frame]
    # En cada frame se fijan los nuevos valores para cada curva y se grafican
    line0.set_data(x0[:frame], t[:frame])
    line1.set_data(x1[:frame], t[:frame])
    line2.set_data(x2[:frame], t[:frame])
    line3.set_data(x3[:frame], t[:frame])
    line4.set_data(x4[:frame], t[:frame])
    text.set_text(f't= {round(t[frame],1)}')
    
    return line0, line1, line2, line3, line4, text

# Creación de la animación, para ello se utiliza la función predefinida FuncAnimation, 
# cuyos parametros construimos con anterioridad
#
animacion = animation.FuncAnimation(fig, update, frames=len(t), init_func=init, interval=200, blit=True)

# Mostrar la animación en el notebook
HTML(animacion.to_jshtml())

# Se guarda en formato gif
#animacion.save(filename="ejemplo2_4.gif", writer="pillow")

```
